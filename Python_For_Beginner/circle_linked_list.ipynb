## 원형 연결 리스트의 일반 구현
위에서는 데이터를 5개로 제한했지만 실제로는 데이터 개수가 제한적이지 않은 경우가 일반적이다.  
메모리의 크기는 무한대로 가정하며 변수 3개를 추가한다.  
* head: 첫 번째 노드를 가리킨다.  
* current: 지금 처리 중인 노드를 가리킨다.  
* pre: 현재 처리 중인 노드의 바로 앞 노드를 가리킨다.

```python
memory = []
head, current, pre = None, None, None
```

```python
## 클래스 함수의 선언 부분 ##

class Node():
    def __init__(self):
        self.data = None
        self.link = None
        
def printNodes(start):
    current = start
    if current.link == None:
        return
    print(current.data, end = ' ')
    while current.link != start: # 현재 작업 위치로 되돌아 올 때까지 빈복
        current = current.link # 다음 링크로 이동하면서
        print(current.data, end=' ') # 현재 데이터를 출력
    print()
        
## 전역 변수 선언 부분 ##
memory = []
head, current, pre = None, None, None
dataArray = ["다현", "정연", "쯔위", "사나", "지효"]

## 메인 코드 부분 ##
if __name__ == "__main__":
    
    node = Node()
    node.data = dataArray[0] # 첫 번째 노드
    head = node
    node.link = head # 자기 자신을 연결
    memory.append(node)
    
    for data in dataArray[1:]: # 두 번째 이후 노드
        pre = node # 현재 작업 노드가 변경 되었으니 pre를 이전 노드로 할당
        node = Node() # 새로운 노드 생성
        node.data = data # 현재 data 할당
        pre.link = node # pre가 보고 있던 노드의 링크, 즉 이전 노드의 링크를 현재 링크로 연결
        node.link = head # 현재 노드의 링크를 head, 즉 첫 번째 데이터에 연결(마지막 데이터니까)
        memory.append(node)
        
    printNodes(head)
```

### 노드 삽입

```python
## 클래스 함수의 선언 부분 ##

class Node():
    def __init__(self):
        self.data = None
        self.link = None
        
def printNodes(start):
    current = start
    if current.link == None:
        return
    print(current.data, end = ' ')
    while current.link != start: # 현재 작업 위치로 되돌아 올 때까지 빈복
        current = current.link # 다음 링크로 이동하면서
        print(current.data, end=' ') # 현재 데이터를 출력
    print()
    
def insertNode(findData, insertData):
    global memory, head, current, pre
    
    if head.data == findData: # 첫 번쩨 노드에 있으면
        node = Node() # 노드 생성
        node.data = insertData # 데이터 생성
        node.link = head # head, 즉 첫 번째 노드에 현재 데이터 link 연결
        last = head # 마지막 노드를 첫 번째 노드로 우선 지정
        while last.link !=head: # 마지막 링크를 찾을 때까지
            last = last.link # last를 다음 노드로 변경
        last.link = node # 마지막 노드의 링크를 새로 생성한 노드에 연결 
        head = node # 현재 생성한 노드를 head로 설정.
        return
    
    current = head
    while current.link != head: # current를 찾을 때 까지 계속 이동
        pre = current
        current = current.link
        if current.data == findData: # 원하는 데이터를 찾으면
            node = Node() # 노드 생성
            node.data = insertData
            node.link = current # 새로 생성한 데이터의 링크를 찾은 데이터에 연결
            pre.link = node # 이전 노드의 링크를 새로 생성한 노드에 연결
            return
    
    node = Node() # 원하는 데이터를 찾지 못해 마지막까지 왔으면
    node.data = insertData # 데이터 생성
    current.link = node # head를 가리키고 있던 link를 현재 노드에 연결
    node.link = head # 생성한 노드의 링크를 head에 연결
    
    
## 전역 변수 선언 부분 ##
memory = []
head, current, pre = None, None, None
dataArray = ["다현", "정연", "쯔위", "사나", "지효"]

## 메인 코드 부분 ##
if __name__ == "__main__":
    
    node = Node()
    node.data = dataArray[0] # 첫 번째 노드
    head = node
    node.link = head # 자기 자신을 연결
    memory.append(node)
    
    for data in dataArray[1:]: # 두 번째 이후 노드
        pre = node # 현재 작업 노드가 변경 되었으니 pre를 이전 노드로 할당
        node = Node() # 새로운 노드 생성
        node.data = data # 현재 data 할당
        pre.link = node # pre가 보고 있던 노드의 링크, 즉 이전 노드의 링크를 현재 링크로 연결
        node.link = head # 현재 노드의 링크를 head, 즉 첫 번째 데이터에 연결(마지막 데이터니까)
        memory.append(node)
        
    printNodes(head)  
    
    insertNode("다현", "회사")
    printNodes(head)
    
    insertNode("사나", "솔라")
    printNodes(head)
    
    insertNode("재남", "문별")
    printNodes(head)
```

### 노드 삭제

```python
## 클래스 함수의 선언 부분 ##

class Node():
    def __init__(self):
        self.data = None
        self.link = None
        
def printNodes(start):
    current = start
    if current.link == None:
        return
    print(current.data, end = ' ')
    while current.link != start: # 현재 작업 위치로 되돌아 올 때까지 빈복
        current = current.link # 다음 링크로 이동하면서
        print(current.data, end=' ') # 현재 데이터를 출력
    print()
    
def deleteNode(deleteData):
    global memory, head, current, pre
    
    if head.data == deleteData: # 첫 번째 노드 삭제
        current = head
        head = head.link
        last = head # 우선 last를 첫 번째 노드로 설정
        while last.link != current: # 마지막 노드를 찾을 때까지
            last = last.link
        last.link = head # 마지막 노드를 찾으면 해당 노드를 "옮긴 head"에 연결
        del(current) # 현재 작업 노드는 연결된 링크가 없음
        return
    
    current = head # 첫 번째 노드가 아니면 우선 current를 첫 번째로 설정
    while current.link != head:
        pre = current
        current = current.link
        if current.data == deleteData: # 지우려는 데이터를 찾으면
            pre.link = current.link # 이전 노드의 링크를 현재 링크가 연결된 노드로 옮김
            del(current)
            return
        
## 전역 변수 선언 부분 ##
memory = []
head, current, pre = None, None, None
dataArray = ["다현", "정연", "쯔위", "사나", "지효"]

## 메인 코드 부분 ##
if __name__ == "__main__":
    
    node = Node()
    node.data = dataArray[0] # 첫 번째 노드
    head = node
    node.link = head # 자기 자신을 연결
    memory.append(node)
    
    for data in dataArray[1:]: # 두 번째 이후 노드
        pre = node # 현재 작업 노드가 변경 되었으니 pre를 이전 노드로 할당
        node = Node() # 새로운 노드 생성
        node.data = data # 현재 data 할당
        pre.link = node # pre가 보고 있던 노드의 링크, 즉 이전 노드의 링크를 현재 링크로 연결
        node.link = head # 현재 노드의 링크를 head, 즉 첫 번째 데이터에 연결(마지막 데이터니까)
        memory.append(node)
        
    printNodes(head) # 초기 상태
    
    deleteNode("다현")
    printNodes(head)
    
    deleteNode("쯔위")
    printNodes(head)
    
    deleteNode("지효")
    printNodes(head)
    
    deleteNode("재남")
    printNodes(head)
```

## 원형 연결 리스트의 응용
* head부터 한 바퀴 돌면서 홀수와 짝수의 개수를 센다.  
* 이 중 개수가 많은 것을 모두 음수로 변경한다.

```python
import random
## 클래스 함수의 선언 부분 ##

class Node():
    def __init__(self):
        self.data = None
        self.link = None
        
def printNodes(start):
    current = start
    if current.link == None:
        return
    print(current.data, end = ' ')
    while current.link != start: # 현재 작업 위치로 되돌아 올 때까지 빈복
        current = current.link # 다음 링크로 이동하면서
        print(current.data, end=' ') # 현재 데이터를 출력
    print()
    
def countOddEven():
    global memory, head, current, pre
    
    odd, even = 0, 0
    if head == None:
        return False # 데이터가 없으면 False 출력
    
    current = head
    while True:
        if current.data % 2 == 0:
            even += 1
        else:
            odd += 1
        if current.link == head: # 마지막이 되면 멈춘다.
            break;
        current = current.link
        
    return odd, even

def makeZeroNumber(odd, even):
    if odd > even:
        reminder = 1
    else:
        reminder = 0
        
    current = head # iteration 돌릴 준비
    while True:
        if current.data % 2 == reminder:
            current.data *= -1
        if current.link == head:
            break;
        current = current.link

## 전역 변수 선언 부분 ##
memory = []
head, current, pre = None, None, None
dataArray = ["다현", "정연", "쯔위", "사나", "지효"]

## 메인 코드 부분 ##
if __name__ == "__main__":
    
    dataArray = []
    for _ in range(7):
        dataArray.append(random.randint(1,100))

    node = Node()
    node.data = dataArray[0] # 첫 번째 노드
    head = node
    node.link = head # 자기 자신을 연결
    memory.append(node)
    
    for data in dataArray[1:]: # 두 번째 이후 노드
        pre = node # 현재 작업 노드가 변경 되었으니 pre를 이전 노드로 할당
        node = Node() # 새로운 노드 생성
        node.data = data # 현재 data 할당
        pre.link = node # pre가 보고 있던 노드의 링크, 즉 이전 노드의 링크를 현재 링크로 연결
        node.link = head # 현재 노드의 링크를 head, 즉 첫 번째 데이터에 연결(마지막 데이터니까)
        memory.append(node)
        
    printNodes(head)
    
    oddCount, evenCount = countOddEven()
    print('홀수 >>', oddCount, '짝수 >>', evenCount)
    
    makeZeroNumber(oddCount, evenCount)
    printNodes(head)
```
